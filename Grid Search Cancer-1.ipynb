```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancerdata=load_breast_cancer()
```

```python
cancerdata
```

```python
data=cancerdata["data"]
```

```python
data
```

```python
dfdata=pd.DataFrame(data,columns=cancerdata['feature_names'])
```

```python
dfdata
```

```python
dftarget=pd.DataFrame(cancerdata["target"],columns=["Target"])
```

```python
dftarget
```

```python
dfcancer=pd.concat([dfdata,dftarget],axis=1)
```

```python
dfcancer
```

```python
X=dfcancer.drop("Target",axis=1)
```

```python
y=dfcancer["Target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,ytrain,ytest=(train_test_split(X,y,test_size=.20))
```

```python
X.shape,Xtrain.shape,Xtest.shape
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
```

```python
svc.fit(Xtrain,ytrain)
```

```python
svc.score(Xtest,ytest)
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(Xtrain)
```

```python
Xtrain_ss=ss.transform(Xtrain)
```

```python
Xtest_ss=ss.transform(Xtest)
```

```python
svc1=SVC()
```

```python
svc1.fit(Xtrain_ss,ytrain)
```

```python
svc1.score(Xtrain_ss,ytrain)
```

```python
svc1.score(Xtest_ss,ytest)
```

# Different Kernels

```python
svcdk=SVC(kernel="rbf")
```

```python
svcdk.fit(Xtrain_ss,ytrain)
```

```python
svcdk.score(Xtest_ss,ytest)
```

```python
svclk=SVC(kernel="linear")
svclk.fit(Xtrain_ss,ytrain)
svclk.score(Xtest_ss,ytest)
```

```python
svcsk=SVC(kernel="sigmoid")
svcsk.fit(Xtrain_ss,ytrain)
svcsk.score(Xtest_ss,ytest)
```

```python
svcpk=SVC(kernel="poly")
svcpk.fit(Xtrain_ss,ytrain)
svcpk.score(Xtest_ss,ytest)
```

# Cross Validation

```python
svc2=SVC()
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
kf1=KFold(n_splits=10)
```

```python
svcpk1=SVC(kernel="poly")
svcdk1=SVC(kernel="rbf")
svcsk1=SVC(kernel="sigmoid")
svclk1=SVC(kernel="linear")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss2=StandardScaler()
```

```python
ss2.fit(X)
```

```python
X_ss=ss2.transform(X)
```

```python
algo=[svcpk1,svcdk1,svcsk1,svclk1]
```

```python
for i in algo:
    result=cross_val_score(i,X_ss,y,cv=kf1)
    print("Model:{},Score:{}".format(i,result.mean()))
```
